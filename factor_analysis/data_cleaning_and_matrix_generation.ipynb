FuzzyMatching workflow
1.   find all the unique OCR'd titles
2.   calculate the three fuzzy matching scores and their average for each potential match among the clean titles
3.   selecte the best match (highest avg fuzzy matching score)

```python
# list the potential matches (clean pub titles)
dic=pd.read_csv('/content/drive/MyDrive/Avant-garde/github_sharing/wordcountAdded_ocred_combined/pub_dictionary.csv')
```

```python
potential_titles_lst=dic['corrected_pub'].unique().tolist()
```

```python
df#(pd.read_csv('/content/drive/MyDrive/Avant-garde/github_sharing/wordcountAdded_ocred_combined/all_raw_pub_count_sort.csv'))
```

```python
#create a new df to save the fuzzy matching predictions
df_fz_pred = pd.DataFrame(columns = potential_titles_lst)#set the columns, each column represents a clean pub
```

```python
df_fz_pred['original_pub']=df['publication'].unique()#set the number of rows, each column represents a raw pub title to match; this column can be deleted after being used for indexing
```

```python
df_fz_pred
```

```python
#set index
df_fz_pred.index = list(df_fz_pred["original_pub"])
```

```python
df_fz_pred
```

```python
# Drop rows that has NaN values on selected columns
df_fz_pred=df_fz_pred.dropna(subset=['original_pub'])
```

```python
len(df_fz_pred) #11558 unique raw pub titles to clean
```

```python
# fuzzy matching: generate three scores
# takes a long time: 11558*83
for cleaned_title in potential_titles_lst:
  avg_score_lst=[]
  for dirty_title in df_fz_pred['original_pub']:
    #print(dirty_title, cleaned_title)
    r1= fuzz.partial_ratio(dirty_title, cleaned_title)
    r2= fuzz.token_sort_ratio(dirty_title,cleaned_title)
    r3= fuzz.token_set_ratio(dirty_title,cleaned_title)
    avg_score=(r1+r2+r3)/3
    avg_score_lst.append(avg_score)
  #print(avg_score_lst)
  df_fz_pred[str(cleaned_title)]= avg_score_lst
```

```python
df_fz_pred
```

```python
#find the best match in each row: the one with the max score
# the highest prob each row
maxValues = df_fz_pred.max(axis = 1)
print(type(maxValues))
print(maxValues)
```

```python
# the corresponding pub index of the highest prob in each role
maxValueIndex = df_fz_pred.idxmax(axis = 1)
print(type(maxValueIndex))
#print("Max values of row are at following columns :")
print(maxValueIndex)
```

```python
fz_dict= pd.DataFrame(maxValueIndex, maxValues).reset_index()

#maxValueIndex.to_frame().join(maxValues.to_frame())
```

```python
fz_dict=pd.concat([maxValueIndex,maxValues],axis=1)
```
